```python
df['date']=pd.to_datetime(df['time'],format = "%H%M%m%d%y")
```

```python
df['date'] = df['date'].apply(lambda x: str(x).replace('-','/'))
```

```python
df['label'] = np.NaN
```

```python
def labels():
    From , to = int(input('Enter Row no. for First value of Row Range: ')),int(input('Enter Row no. for Second value of Row range: '))
    while From and to not in range(len(df)):
        print(f'Please enter value of From and to between 0-{len(df)} ')
        From , to = int(input('Enter Row no. for First value of Row Range: ')),int(input('Enter Row no. for Second value of Row range: '))
    label = int(input('Enter label between 0-9: '))
    
    while label not in range(10):
            
        print('Please enter label between 0-9 ')
        label = int(input('Enter label between 0-9: '))
    df.loc[From:to+1, 'label'] =label
        
labels()    
```
